### Numpy basics

1\. Find the row, column and overall means for the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python
import numpy as np

m = np.arange(12).reshape((3,4))
print("Original matrix:",'\n',m,'\n')

#Row mean
m_row = m.mean(axis=1)
print("Row mean:",'\n',m_row,'\n')

#Column mean
m_column = m.mean(axis=0)
print("Column mean:",'\n',m_column,'\n')

#Overall mean
m_mean = m.mean()
print("Overall mean:",'\n',m_mean)
````

2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

#Outer product of u and v:

#1) Using the function outer

product1 = np.outer(u,v)

print("The outer product using the function outer in numpy is:",'\n',product1,'\n')

#2) Using a nested for loop

product2 = np.eye(4) # 4x4 unitary matrix

for i in range(0,4):
    for j in range(0,4):
        product2[i,j] = u[i]*v[j] #filling the matrix element by element 

print("The outer product using a nested for loop is:",'\n',product2,'\n')

#3) Using numpy broadcasting operations

u = u[:,np.newaxis] # adding a dimension along axis 1
v = v[np.newaxis,:] # adding a dimension along axis 0

product3 =u*v

print("The outer product using numpy broadcasting operations is:",'\n',product3,'\n')
````

3\. Create a 10 by 6 matrix of random uniform numbers. Set all rows with any entry less than 0.1 to be zero

Hint: Use the following numpy functions - np.random.random, np.any as well as Boolean indexing and the axis argument.

````python
#0 by 6 matrix of random uniform numbers
my_matrix = np.random.random((10,6)) 

print("10 by 6 matrix of random uniform numbers:",'\n',my_matrix,'\n')

#Set all rows with any entry less than 0.1 to be zero

mask = (my_matrix < 0.1).any(axis=1) #Test whether any matrix element along the rows (axis=1) evaluates to True.
my_matrix[mask, :] = 0 #Where mask is True, set all the elements in this row equal to 0

print("Resulted matrix:",'\n',my_matrix,'\n')
````

4\. Use np.linspace to create an array of 100 numbers between 0 and 2π (includsive).

  * Extract every 10th element using slice notation
  * Reverse the array using slice notation
  * Extract elements where the absolute difference between the sine and cosine functions evaluated at that element is less than 0.1
  * Make a plot showing the sin and cos functions and indicate where they are close

````python
from math import pi

number = np.linspace(0, 2*pi, 100) # start, end, num-points
print("The array of 100 numbers between 0 and 2π is:",'\n',number,'\n')

#1) Extract every 10th element using slice notation

number_10 = number[::10]
print("Numbers every 10th element are:",'\n',number_10,'\n')

#2) Reverse the array using slice notation

number_reverse = number[::-1]
print("The reverse array is:",'\n',number_reverse,'\n')

#3) Extract elements where the absolute difference between sine and cosine functions evaluated at that element is less than 0.1

mask = np.abs(np.sin(number)-np.cos(number)) < 0.1
filtered_number = number[mask]
print("Filtered array is:",'\n',filtered_number,'\n')

#4) Plot that shows the sin and cos functions

import matplotlib.pyplot as plt 

%matplotlib inline 

plt.plot(number, np.sin(number),label='sine')
plt.plot(number, np.cos(number),label='cosine')
plt.scatter(filtered_number, np.sin(filtered_number)) #Indicate where sin and cosine are close
plt.legend()
````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anto-diagonal (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagnoal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
#Matrix that shows the 10 by 10 multiplication table

N = 10
x = np.arange(1, N+1)
x_t = np.array([x]).T

matrix = x * x_t
print("10 by 10 multiplication table is:",'\n',matrix,'\n')

#1) Find the trace of the matrix

trace = matrix.trace()
print("The trace is:",'\n',trace,'\n')

#2) Extract the anti-diagonal 

flipped_matrix = np.fliplr(matrix)
anti_diagonal = flipped_matrix.diagonal()
print("The anti-diagonal is:",'\n',anti_diagonal,'\n')

#3) Extract the diagonal offset by 1 upwards

new_matrix = matrix.diagonal(offset=1)
print("The diagonal offset by 1 upwards is",'\n',new_matrix,'\n')
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
#Use broadcasting to create a grid of distances

#1) Construct a 2D grid of distances among each city along Route 66

Chicago = (0,0)
Springfield = (198,0)
SaintLouis = (303 ,0)
Tulsa = (736,0)
Oklahoma = (871,0)
Amarillo = (1175,0)
SantaFe = (1475,0)
Albuquerque = (1544,0)
Flagstaff = (1913,0)
LosAngeles = (2448,0)

points = np.array([Chicago, Springfield, SaintLouis, Tulsa, Oklahoma, Amarillo, SantaFe, Albuquerque, Flagstaff, LosAngeles])

distance = np.sum( abs(points[None,:] - points[:, None]) ,-1) #broadcasting

print("2D grid of distances among each city along Route 66 is:",'\n',distance,'\n')

#2) Conversion from miles to km 

conversion_factor = 0.62137119
distance_km = np.round(distance/conversion_factor,2)

print("Grid of distances in km is:",'\n',distance_km)
            
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import time

start_time = time.time()

N = 99
nums = np.arange(0,N+1) #array with number from 0 to N

#1) Constract a shape (100,) boolean array

mask = np.full(N+1, False, dtype=bool) #Initially, all the elements are False 

#2) Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element

index =[i for i in range(len(nums)) if sum([nums[i]%val==0 for val in range(2,nums[i]-1)])==0]

for j in range(2,len(index)):
    for i in range(2,100):
        if index[j] == nums[i]:
            mask[i]=True  #If the number is prime, set to True the corresponding element in the mask 

#3) Apply the mask to obtain an array of ordered prime numbers

print("The prime numbers from 0 to", N, "are:",'\n',nums[mask],'\n')

#4) Performance

time = time.time()-start_time
print("The simulation runs in:",'\n',time,"s",'\n')
#If we increase N, the simulation takes more time to be executed

#5) Implement the optimization suggested in the sieve of Eratosthenes

#A value in A[i] will be False if i is not a prime, else True. 

m = int(np.sqrt(N)) 
A = np.ones(N, dtype=bool)#array of Boolean values, indexed by integers 2 to N initially all set to true.
A[:2] = False  # 0 and 1 are not primes

for i in range(2, m): #Initially, p = 2 is the smallest prime number
    if A[i] == False:
        continue
    A[i*i::i] = False

prime_numbers = np.nonzero(A)[0] #np.nonzero(A) find the indices of A, where a condition is True
print("The prime numbers using the sieve of Eratosthenes are:",'\n',prime_numbers)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
from math import sqrt
import random
import matplotlib.pyplot as plt

#1) Take 1000 walkers and let them walk for 200 steps
n_walkers = 1000
steps = 200

#2) 2D array of size walkers x steps with values -1 or 1
matrix = np.zeros((n_walkers,steps))

for i in range(n_walkers):
    matrix[i][0] == 0
    matrix[i][1] = 2*np.random.randint(0,2)-1
    
for i in range(n_walkers):
    for j in range(1,steps):
        n = 2*np.random.randint(0,2)-1
        walk = matrix[i][j-1]
        matrix[i][j] = (n+walk)
        
print("The matrix related to the random walking distance is:",'\n',matrix)

#3) Actual walking distances for each walker
d = matrix.cumsum(axis=1)
print("The actual ealking distance for each walker:",'\n',abs(d))

#4) Square of that 2D array
d = d**2
print("The squared matrix:",'\n',d) 

#5) Mean of the squared distances at each step
square_distance = d.mean(axis=0)
print("The mean of the squared distances:",'\n',square_distance)

#6) Plot of the average distances as a function of time
time = [i for i in range(steps)]
av_mean = [sqrt(i) for i in square_distance]

%matplotlib inline
plt.plot(time,av_mean,label="average distance")
plt.xlabel("time")
plt.legend()
plt.show()


#I did not get what I expected. Indeed it is a diffusion process, so I expected a curve like a square root curve that 
#decrease with the time. 
````

Do you feel there is some evident correlation here? [Studies](https://www.enr.gov.nt.ca/en/services/lynx/lynx-snowshoe-hare-cycle) tend to believe so.

````python
import matplotlib.pyplot as plt

#1) Check the content by looking within the file
f = open('populations.txt','r')
file_contents = f.read()

#2) Load the data (use an appropriate numpy method) into a 2D array
array2d = np.loadtxt('populations.txt').astype('int')

#3) Create arrays out of the columns, the arrays being (in order): year, hares, lynxes, carrots
year = np.array([array2d[i][0] for i in range(len(array2d))])
hares = np.array([array2d[i][1] for i in range(len(array2d))])
lynxes = np.array([array2d[i][2] for i in range(len(array2d))])
carrots = np.array([array2d[i][3] for i in range(len(array2d))])

#4) Plot the 3 populations over the years
xs = year
plt.plot(xs, hares, label="hares")
plt.plot(xs, lynxes, label="lynxes")
plt.plot(xs, carrots, label="carrots")
plt.xlabel("year")
plt.legend()
plt.show()

#5) Compute the main statistical properties of the dataset (mean, std, correlations, etc.)
populations = array2d[:, 1:]
mean_data = populations.mean(axis=0)
print ("Mean (hares, lynxes, carrots):", np.round(mean_data,2))
std_data = populations.std(axis=0)
print ("Std (hares, lynxes, carrots):", np.round(std_data,2))
corr1 = np.corrcoef(hares,lynxes)[0,1]
print ("Correlation between hares and lynxes:",np.round(corr1,2))
corr2 = np.corrcoef(hares,carrots)[0,1]
print ("Correlation between hares and carrots:",np.round(corr2,2))
corr3 = np.corrcoef(carrots,lynxes)[0,1]
print ("Correlation between carrots and lynxes:",np.round(corr3,2),'\n')

#6) Which species has the largest population for each year?
maximum = array2d.max(axis=1)
name = []
for i in range(len(maximum)):
    if maximum[i] == hares[i]:
        name.append("hares")
    if maximum[i] == lynxes[i]:
        name.append("lynxes")
    else:
        name.append("carrots")

print("Species with the largest population for each year:")
for i in range(len(year)):
    print(year[i],name[i])
````
